# 구 별로 데이터 추출

```python
gu_list = ['종로구', '중구', '용산구', '성동구', '광진구', '동대문구', '중랑구', '성북구', '강북구',
       '도봉구', '노원구', '은평구', '서대문구', '마포구', '양천구', '강서구', '구로구', '금천구',
       '영등포구', '동작구', '관악구', '서초구', '강남구', '송파구', '강동구']
```

# 동 별로 데이터 추출

```python
df_dong = cost_2019_06.loc[cost_2019_06['자치구명']=='종로구']
df_dong.tail()
```

```python
df_dong['법정동명'].unique()
```

```python
# 동 이름 리스트 생성
dong_list = df_dong['법정동명'].unique().tolist()
len(dong_list)
```

# 전세구분
### 전세가, 월세 매물이 없을 수 있다는 것도 고려하기

```python
df_year = df_dong.loc[df_dong['전월세구분']=='전세']
df_year.tail(20)
```

```python
# len(df_year['법정동명'].unique())


# df_year[df_year['보증금'].isnull()]
# cost_2019_06.drop([175], inplace=True)
# cost_2019_06.tail()
```

```python
dong_cost_year = []


for n in dong_list:
    df_tmp = df_year.loc[df_year['법정동명']==n] 
    dong_cost_year.append(df_tmp['보증금'].mean())
    
    
dong_cost_year
```

```python
# 소수점 둘째자리까지 반올림
dong_cost_year = np.round(dong_cost_year, 2)
dong_cost_year = dong_cost_year.tolist()
dong_cost_year
```

# 월세구분

```python
df_month = df_dong.loc[df_dong['전월세구분']=='월세']
df_month.tail(20)
```

```python
df_dong.loc[(df_dong['전월세구분']=='월세') & (df_dong['법정동명']=='필운동')]
```

```python
# 월세 보증금 평균가격

dong_cost_month_bo = []


for n in dong_list:
    df_tmp = df_month.loc[df_month['법정동명']==n] 
    
    
    dong_cost_month_bo.append(df_tmp['보증금'].mean())
    
    
dong_cost_month_bo
```

```python
dong_cost_month_bo = np.round(dong_cost_month_bo, 2)
dong_cost_month_bo = dong_cost_month_bo.tolist()
dong_cost_month_bo
```

```python
# 월세 임대료 평균가격

dong_cost_month_im = []


for n in dong_list:
    df_tmp = df_month.loc[df_month['법정동명']==n] 
    
    
    dong_cost_month_im.append(df_tmp['임대료'].mean())
    
    
dong_cost_month_im
```

```python
dong_cost_month_im = np.round(dong_cost_month_im, 2)
dong_cost_month_im = dong_cost_month_im.tolist()
dong_cost_month_im
```

```python
# 동 별 전세평균, 월세보증금평균, 월세임대료평균


dong_cost_final = pd.DataFrame({'동':dong_list,
                              '전세평균':dong_cost_year,
                             '월세보증금평균':dong_cost_month_bo,
                             '월세임대료평균':dong_cost_month_im}
                            )

dong_cost_final

```

```python
# 소수점 둘째자리까지

pd.options.display.float_format = '{:.2f}'.format
```
