# 서울특별시 전월세가 정보

```python
# https://data.seoul.go.kr/dataList/datasetView.do?infId=OA-15549&srvType=S&serviceKind=1&currentPageNo=1

cost_2019_06 = pd.read_csv('../project_data/cost/서울특별시 전월세가 정보.csv', encoding='utf-8')
# cost_2019_06.head()
```

```python
# 자치구명 = 구 이름
# cost_2019_06['자치구명'].unique()
```

```python
# null 값 없애줌
cost_2019_06[cost_2019_06['자치구명'].isnull()]
# cost_2019_06.drop([175], inplace=True)
# cost_2019_06.tail()
```

```python
view_columns = ['자치구명', '법정동명', '전월세구분', '보증금', '임대료', '임대건물명']
cost_2019_06 = cost_2019_06[view_columns].copy()
cost_2019_06.head()
```

```python
# cost_2019_06 = cost_2019_06[cost_2019_06['임대건물명']=='다세대/연립']
# cost_2019_06['임대건물명'].value_counts()
```

# 구 별로 데이터 추출

```python
gu_list = cost_2019_06['자치구명'].unique().tolist()
len(gu_list)
```

# 동 별로 데이터 추출

```python
dong_total_list = cost_2019_06['법정동명'].unique().tolist()
len(dong_total_list)
```

# 전세구분
### 전세가, 월세 매물이 없을 수 있다는 것도 고려하기

```python
df_year = cost_2019_06.loc[cost_2019_06['전월세구분']=='전세']
# df_year.tail(20)
```

```python
# len(df_year['법정동명'].unique())


# df_year[df_year['보증금'].isnull()]
# cost_2019_06.drop([175], inplace=True)
# cost_2019_06.tail()
```

```python
dong_cost_year = []


for n in dong_total_list:
    df_tmp = df_year.loc[df_year['법정동명']==n] 
    dong_cost_year.append(df_tmp['보증금'].mean())
    
    
len(dong_cost_year)
```

```python
# 소수점 둘째자리까지 반올림
dong_cost_year = np.round(dong_cost_year, 2)
dong_cost_year = dong_cost_year.tolist()
# dong_cost_year
```

# 월세구분

```python
df_month = cost_2019_06.loc[cost_2019_06['전월세구분']=='월세']
df_month.head()
```

```python
# 월세 보증금 평균가격

dong_cost_month_bo = []


for n in dong_total_list:
    df_tmp = df_month.loc[df_month['법정동명']==n] 
    
    
    dong_cost_month_bo.append(df_tmp['보증금'].mean())
    
    
len(dong_cost_month_bo)
```

```python
dong_cost_month_bo = np.round(dong_cost_month_bo, 2)
dong_cost_month_bo = dong_cost_month_bo.tolist()
len(dong_cost_month_bo)
```

```python
# 월세 임대료 평균가격

dong_cost_month_im = []


for n in dong_total_list:
    df_tmp = df_month.loc[df_month['법정동명']==n] 
    
    
    dong_cost_month_im.append(df_tmp['임대료'].mean())
    
    
# dong_cost_month_im
```

```python
dong_cost_month_im = np.round(dong_cost_month_im, 2)
dong_cost_month_im = dong_cost_month_im.tolist()
# dong_cost_month_im
```

## 구 리스트 생성

```python
gu_total_list = []


for n in gu_list:
#     print(n)
    df_dong = cost_2019_06.loc[cost_2019_06['자치구명']==n]
    
#     print(df_dong['법정동명'].unique())
#     print(len(df_dong['법정동명'].unique()))
    
    for i in range(1, (len(df_dong['법정동명'].unique()))+1):
        gu_total_list.append(n)

len(gu_total_list)    
```

```python
del gu_total_list[0]
len(gu_total_list)
```

# 데이터프레임 생성

```python
dong_cost_final = pd.DataFrame({'구':gu_total_list,
                                '동':dong_total_list,
                              '전세평균':dong_cost_year,
                             '월세보증금평균':dong_cost_month_bo,
                             '월세임대료평균':dong_cost_month_im}
                            )

dong_cost_final.tail(50)
```

```python
dong_cost_final.index.name = 'eno'
```
